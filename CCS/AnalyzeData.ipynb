```python
import cv2
import matplotlib.pyplot as plt
import os
from tqdm import tqdm

# Function to get the resolution of an image
def get_image_resolution(image_path):
    # Read the image
    image = cv2.imread(image_path)
    
    # Check if the image was successfully loaded
    if image is None:
        print(f"Error: Unable to load image at {image_path}")
        return None
    
    # Get the resolution of the image
    height, width = image.shape[:2]
    return width, height

# Function to plot the histogram of resolutions
def plot_resolution_histogram(image_paths):
    # List to store resolutions
    resolutions = []
    
    # Iterate through image paths and get resolutions
    for image_path in tqdm(image_paths):
        resolution = get_image_resolution(image_path)
        if resolution is not None:
            resolutions.append(resolution)
    
    # Separate width and height for histogram plotting
    widths = [res[0] for res in resolutions]
    heights = [res[1] for res in resolutions]
    
    # Plot histogram for widths
    plt.figure()
    plt.hist(widths, bins=20, alpha=0.5, label='Widths')
    plt.hist(heights, bins=20, alpha=0.5, label='Heights')
    plt.title('Histogram of SCC')
    plt.xlabel('Pixels')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.show()

# Example usage
# Directory containing your images
image_directory = 'Data/SCC'
# Get list of all image files in the directory
image_paths = [os.path.join(image_directory, f) for f in os.listdir(image_directory) if f.endswith(('jpg', 'jpeg', 'png'))]

plot_resolution_histogram(image_paths)
```
